##### Principais caracteristicas do sinal sonoro
***

Estas sao as caracteristicas extraidas pelo *PyAudioAnalisys*. As features são extraidas em janelas de curta duraçao (Short Window), da ordem de milisegundos e podem ser representadas em pela media e desvio padrao de janelas de tempo maiores, determinadas Midterm Windows ou segmentos. 

| Feature ID |    Feature Name    |                                                                         Description                                                                         |
|:----------:|:------------------:|:-----------------------------------------------------------------------------------------------------------------------------------------------------------:|
|      1     | Zero Crossing Rate |                                      The rate of sign-changes of the signal during the duration of a particular frame.                                      |
|      2     |       Energy       |                                     The sum of squares of the signal values, normalized by the respective frame length.                                     |
|      3     |  Entropy of Energy |                            The entropy of sub-frames' normalized energies. It can be interpreted as a measure of abrupt changes.                            |
|      4     |  Spectral Centroid |                                                            The center of gravity of the spectrum.                                                           |
|      5     |   Spectral Spread  |                                                          The second central moment of the spectrum.                                                         |
|      6     |  Spectral Entropy  |                                             Entropy of the normalized spectral energies for a set of sub-frames.                                            |
|      7     |    Spectral Flux   |                            The squared difference between the normalized magnitudes of the spectra of the two successive frames.                            |
|      8     |  Spectral Rolloff  |                                 The frequency below which 90% of the magnitude distribution of the spectrum is concentrated.                                |
|    9-21    |        MFCCs       |  Mel Frequency Cepstral Coefficients form a cepstral representation  where the frequency bands are not linear but distributed according to  the mel-scale.  |
|    22-33   |    Chroma Vector   | A 12-element representation of the spectral energy where the bins  represent the 12 equal-tempered pitch classes of western-type music  (semitone spacing). |
|     34     |  Chroma Deviation  |                                                    The standard deviation of the 12 chroma coefficients.                                                 |


* **Short-term feature extraction**: this is implemented in function feature_extraction() of the ShortTermFeatures.py file. It splits the input signal into short-term widnows (frames) and computes a number of features for each frame. This process leads to a sequence of short-term feature vectors for the whole signal.
* **Mid-term feature extraction**: In many cases, the signal is represented by statistics on the extracted short-term feature sequences described above. Towards this end, function mid_feature_extraction() from the MidTermFeatures.py file extracts a number of statistcs (e.g. mean and standard deviation) over each short-term feature sequence.

The total number of short-term features implemented in pyAudioAnalysis is 34. In addition, the delta features are optionally computed (they are by default enabled, but can be disabled by setting the deltas argument in feature_extraction() to false). So, the total number of short-term features, including the deltas is 64.

**Janelas de curto prazo**

Considerando um sinal sonoro de tempo **T**, podemos dividi-lo em janelas curtas, as *ShortWindows* ou *frames*, de forma a extrair caraterísticas delas de forma sistemática. 

Os tamanhos das janelas curtas, de acordo com Giannakopoulos (20xx), tem valores típicos entre 10 e 100ms, a depender da aplicação.

Outro parâmetro utilizado é o **passo** ou **step**, que determina a distância entre uma janela e a seguinte. Caso o **step** < **frame**, há sobreposição de janelas, conforme mostrado na figura. Para o caso onde não há sobreposição, o **step** e o **frame** são iguais.

Caso o **step** < **frame**, então há sobreposição. Por exemplo, com **step** = 10ms e um **frame** = 40ms, haverá 75% de sobreposição.

Supondo que seja possivel dividir o sinal de tempo **T** em **N** frames. Para cada **frame** são extraídos um conjunto de caracteristicas, que podem ser de três domínios distintos:

* Domínio do tempo: Como amplitude máxima, energia e outras.
* Dómínio da frequência: São calculadas após a transformação do final por meio da FFT, por exemplo, centroide espectral.
* Domínio Cepstral: é uma transformação logarítimca aplicada aos componentes do domínio da frequência, cuja idéia principal é representar os componentes do som dando importância maior aos que mais sensibilizam o ouvido humano.

**Janelas de médio prazo (segmentação)**

Como representar diversos sinais sonoros de tempos arbitrários em um único dataset? 

Uma solução seria identificar a duração do maior sinal do *dataset* e toma-lo como padrão. Todas as *features*  calculadas nas *short windows* seriam concatenadas de forma que elas formassem uma única entrada (uma única linha) neste *dataset* representando este sinal. Os demais sinais (que são menores que ele) seriam adicionados ao *dataset* da mesma forma, apenas realizando o *padding* com zeros por conta da duração menor.

O problema desta abordagem é que acarretará uma alta dimensionalidade do dataset, as *features* serão dependentes do tempo em que ocorreram, e na adição de um sinal de áudio com duração superior ao maior presente na base, alteram-se a configuração de *features*

Diante do exposto, a forma mais adotada de representação tem sido calcular as janelas de curta duração e agrupa-las  em segmentos, onde se calcula a média e o desvio-padrão das features presentes no intervalo definido como janela de médio prazo (ou seguimento) 

Por exemplo:
* Sinal sonoro com duração de 2,5s
* Janelas de curto prazo de 50ms, ou 0,005s. Considerando o *step* igual a duração da janela, ou seja, sem sobreposição.
* Janelas de médio prazo (segmentos) de 1s.
***
* Existem então 2,5s / 0,05s = 50 janelas de curto prazo.
* Existem 2,5s / 1s = 2,5 duas janelas e meia (três segmentos de 1s). Onde os dois primeiros segmentos tem 20 janelas de curto prazo e o ultimo segmento tem 10 janelas de curto prazo. 

As médias e os desvios das *features* das janelas de curto prazo são calculadas e representam os valores dos segmentos. Neste caso, os 20 conjuntos de *features* extraidas das janelas de curto prazo são utilizados.

Ao final, o sinal como um todo pode ser presentado pela média e desvio dos valores obtidos nos segmentos, porém sem muita utilidade prática.

```python
Features_iniciais = ['zcr_mean', 'energy_mean', 'energy_entropy_mean', 
 'spectral_centroid_mean', 'spectral_spread_mean', 'spectral_entropy_mean',
 'spectral_flux_mean', 'spectral_rolloff_mean', 'mfcc_1_mean']

data_std = df_longterm[Features_iniciais]
data_std = preprocessing.scale(data_std)
data_std = pd.DataFrame(data = data_std, columns = Features_iniciais)
data_std['target'] = df_longterm['Classe_supervisionada']
```

```python
from h2o.estimators import H2OSupportVectorMachineEstimator
# utilizar h2o autoML no proximo passo


df = data_std.copy(deep = True)
df['target'] = pd.Series(df['target'])

df.drop(['target'], axis=1, inplace=True)
df = h2o.H2OFrame(data_std) # tranforma um pandasDf em h2oDf


df['target'] = df['target'].asfactor()

# set the predictor and response columns
predictors = ['zcr_mean', 'energy_mean', 'energy_entropy_mean', 
 'spectral_centroid_mean', 'spectral_spread_mean', 'spectral_entropy_mean',
 'spectral_flux_mean', 'spectral_rolloff_mean', 'mfcc_1_mean']
response_col = "target"

# split into train and testing sets
train, test = df.split_frame(ratios = [0.6], seed = 1)

# Initialize model
SVM_model = H2OSupportVectorMachineEstimator()

SVM_model.train(predictors, response_col, training_frame = train)

# predict using the model and the testing dataset
perf = SVM_model.model_performance(test)

# View a summary of the prediction
perf
```
